```python
path="EEGwithEDA.csv"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import statsmodels.api as sm
```

```python
#脳波を羅列する
#実際にはobs, time, Delta, Theta, Alpha1, Alpha2, Beta1, Beta2, Gamma1, Gamma2, Attention, Meditation, Derived, totPwr, CLASSを並べる
```

```python
columns=["obs", "time", "Delta", "Theta", "Alpha1", "Alpha2", "Beta1", "Beta2", "Gamma1", "Gamma2", "Attention", "Meditation", "Derived", "totPwr", "class" ]
sampling_rate=512
df=None

reader = pd.read_csv(path, encoding="SHIFT_JIS",header=None,names=columns)
reader=reader.drop(0,axis=0)#一行目はいらない
```

```python
print(reader)
```

```python
##Deltaを変換する
df_D=[]
#Deltaを格納
df_D.append(reader.Delta)
print(df_D)
```

```python
#objectになっているのでintに変換
df_D=np.array(df_D,dtype=np.int64)
#二次元配列はグラフにしてくれないので１次元にする
df_D=df_D.flatten()
```

```python
#中身の確認
print(df_D)
```

```python
#グラフにする
plt.plot(df_D)
```

```python
##timeも変換してDeltaと一緒に見えるようにする
df_t=[]
df_t.append(reader.time)
df_t=np.array(df_t,dtype=np.float64)
df_t=df_t.flatten()
##atten
```

```python
plt.plot(df_t,df_D)
```
